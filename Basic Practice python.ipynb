## pie chart

```python
import matplotlib.pyplot as plt
 
```

```python
food_menu = ['pizza', 'Burger', 'Hotdog', 'chicken', 'gimbap']
count = [20, 13, 40, 30, 50]

plt.pie(count, labels=food_menu, autopct='%1.2f%%', shadow=True, colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown', 'red'], startangle=90)
plt.title("Restaurant menu availability", fontsize=30, color='darkcyan')
plt.show()
```

```python
food_menu = ['pizza', 'Burger', 'Hotdog', 'chicken', 'gimbap']
count = [20, 13, 40, 30, 50]

explode = (0, 0, 0, 0, 0.1)
pairs = zip(food_menu, count)
pairs = sorted(list(pairs))
food_menu, count = zip(*pairs)

plt.pie(count, labels=food_menu, explode=explode, autopct='%.2f%%', counterclock=False, shadow=False)
plt.show()
```

## subplots and multiple windows

```python
x = np.arange(-100,100,3)
y1 = 0.5 * x 
y2 = np.sin(x)

ax1 = plt.subplot(212)
ax2 = plt.subplot(211)

ax1.plot(x,y1)
ax2.plot(x,y2)
plt.show()
```

```python
x = np.arange(-100,100,3)
y1 = 0.5 * x 
y2 = np.sin(x)

ax1 = plt.subplot(222)
ax2 = plt.subplot(221)

ax1.plot(x,y1)
ax2.plot(x,y2)
plt.tight_layout()  # to not coincide or overwrite
plt.show()
```

```python
x = np.arange(-100,100,3)
y1 = 0.5 * x 
y2 = np.sin(x)

ax1 = plt.subplot(222)
ax2 = plt.subplot(221)
ax3 = plt.subplot(223)
ax4 = plt.subplot(224)

ax1.plot(x,y1, 'yellow')
ax2.plot(x,y2, 'gray')
ax3.plot(x,y2, 'r>')
ax4.plot(x,y2, 'g')
plt.tight_layout()  # to not coincide or overwrite
plt.show()
```

```python
x = np.arange(-100,100,3)
y1 = 0.5 * x 
y2 = np.sin(x)

plt.figure(1)
ax1 = plt.subplot(222)
ax2 = plt.subplot(221)
ax1.plot(x,y1, 'yellow')
ax2.plot(x,y2, 'gray')

plt.figure(2)
plt.plot(x,y1)

plt.tight_layout()  # to not coincide or overwrite
plt.show()
```

## bar charts

```python
style.use('classic')
python = (56, 43, 76, 78)
sql = (34, 95, 78, 98)
power_BI = (23, 42, 15, 90)
Tableau = (55, 66, 77, 88)

people = ['lavanya', 'karu', 'nama', 'santhu']

index = np.arange(4)

plt.bar(index, python, width=0.3, label="python")
plt.bar(index + 0.2, sql, width=0.3, label="sql")
plt.bar(index + 0.4, power_BI, width=0.3, label="power_BI")
plt.bar(index + 0.6, Tableau, width=0.3, label="Tableau")

plt.title("skill levels")
plt.xlabel("person")
plt.ylabel("skill level")
plt.xticks(index+0.2, people)
plt.legend(loc='upper left')
plt.ylim(0, 150)
plt.show()
```
